```python
# Regular Expressions: pattern matching, pattern replacement, and character splitting
# https://docs.python.org/3/howto/regex.html documentation for regex for practice

# installing libraries
import numpy as np
import pandas as pd
from sklearn.neighbors import DistanceMetric
from math import radians
import sys
from math import dist
from geopy import distance
```

```python
# reading accident file
accidents = pd.read_csv('US_accidents_data_subset_MA_feb28.csv')
```

```python
# reading fire department file
boston_firedpt = pd.read_csv('Fire_Departments.csv')
```

```python
boston_firedpt.head(10)
```

```python
# creating boston subset
df_accidents_boston = accidents[accidents["City"]=="Boston"]
df_accidents_boston
```

```python
df_accidents_boston2 = df_accidents_boston[['ID', 'Severity', 'Start_Time', 'End_Time', 'Start_Lat', 'Start_Lng', 'Distance(mi)', 'Description', 'Number', 'Street', 'City', 'Zipcode']]
df_accidents_boston2
```

```python
firedpt_subset2 = boston_firedpt[boston_firedpt["XCOORD"]!=0]
firedpt_subset2 = boston_firedpt[boston_firedpt["YCOORD"]!=0]
firedpt_subset2.dropna(axis=0,inplace=True)
firedpt_subset2
```

```python
# Modify XCOORD and YCOORD as coordinates. You can adapt the following, which adds a decimal after 2 places:
firedpt_subset2['XCOORD'] = firedpt_subset2['XCOORD'].apply(lambda x: x / 10 ** (len((str(x))) - 2))*100
firedpt_subset2['YCOORD'] = firedpt_subset2['YCOORD'].apply(lambda x: x / 10 ** (len((str(x))) - 2))*100
firedpt_subset2
```

```python
# Extract the coordinates from both dataframes as dictionaries. The following is an example:
boston_coordinates = df_accidents_boston2[["ID","Start_Lat","Start_Lng"]]
boston_dict = boston_coordinates.set_index('ID').T.to_dict('list')

firedept_coordinates = firedpt_subset2[['OBJECTID_1','XCOORD','YCOORD']]
firedept_dict = firedept_coordinates.set_index('OBJECTID_1').T.to_dict('list')
```

```python
boston_dict
```

```python
firedept_dict
```

```python
# Combine the dataframes and add a new column “minimum distance” that calculates the 
# Euclidean distance between the coordinates
# https://docs.python.org/3/library/math.html

XCOORD = []
YCOORD = []
Min_Distance = []

for [x,y] in boston_dict.values():
    min_distance=sys.maxsize
    for [p,q] in firedept_dict.values():
        res = dist([x,y], [p,q])
        if res < min_distance:
            min_distance = res
            key1 = {i for i in firedept_dict if firedept_dict[i]==[p,q]}
            p_value = p
            q_value = q
    XCOORD.append(p_value)
    YCOORD.append(q_value)
    Min_Distance.append(min_distance)
    key2 = {i for i in boston_dict if boston_dict[i]==[x,y]}
    
df_accidents_boston["XCOORD"] = XCOORD
df_accidents_boston["YCOORD"] = YCOORD
df_accidents_boston["Minimum Distance"] = Min_Distance

df_accidents_boston2["XCOORD"] = XCOORD
df_accidents_boston2["YCOORD"] = YCOORD
df_accidents_boston2["Minimum Distance"] = Min_Distance
```
